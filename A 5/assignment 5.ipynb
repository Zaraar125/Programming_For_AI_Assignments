```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel("data.xlsx")
test=pd.read_excel("test.xlsx")
test1=pd.read_excel("test.xlsx")
```

```python
work={'Unemployed':0,'Self Employed':1,'Public Sector':2,'Private sector':3}
gend={'Male':0,'Female':1}
deg={'None':0,'Other':1,'Degree':2,'GCSE':3,'A Level':4,'Masters':5,'PhD':6}
reg
```

```python
data["WorkType"]=data["WorkType"].map(work)
data["Gender"]=data["Gender"].map(gend)
data["Education"]=data["Education"].map(deg)

test["WorkType"]=test["WorkType"].map(work)
test["Gender"]=test["Gender"].map(gend)
test["Education"]=test["Education"].map(deg)
```

```python
answer=[]
z=0
for i in range(len(test)):
    #a=data[(data["Education"]==test.iloc[i][4]) & (data["WorkType"]==test.iloc[i][5])& (data["Gender"]==test.iloc[i][6]) & (data["Region"]==test.iloc[i][7])]
    a=data[data["Region"]==test.iloc[i][7]]
    d=[]
    for j in range(len(a.iloc[:])):
        z=float(((test.iloc[i][0]-a.iloc[j][0])**2+(test.iloc[i][1]-a.iloc[j][1])**2+(test.iloc[i][2]-a.iloc[j][2])**2+(test.iloc[i][3]-a.iloc[j][3])**2+(test.iloc[i][4]-a.iloc[j][4])**2+(test.iloc[i][5]-a.iloc[j][5])**2+(test.iloc[i][6]-a.iloc[j][6])**2)**0.5)
        d.append(float(z))
    a["distance"]=d
    a=a.sort_values("distance",ascending=True)
    a.reset_index(inplace=True,drop=True)
    answer.append(int(a.iloc[0]["Salary"]))
new_ans=[]
for i in answer:
    if i>35000:
        new_ans.append(1)
    else :
        new_ans.append(0)
test1["Salary"]=answer
test1["Bool Salary"]=new_ans
```

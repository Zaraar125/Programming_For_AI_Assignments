```python
bop_mean=[]
abl_mean=[]
hbl_mean=[]
ubl_mean=[]
for i in bop["new_year"].unique():
    bop_mean.append(float(bop[bop["new_year"]==i]["Close"].mean()))
for i in abl["new_year"].unique():
    abl_mean.append(float(abl[abl["new_year"]==i]["Close"].mean()))
for i in ubl["new_year"].unique():
    ubl_mean.append(float(ubl[ubl["new_year"]==i]["Close"].mean()))
for i in bop["new_year"].unique():
    hbl_mean.append(float(hbl[hbl["new_year"]==i]["Close"].mean()))
```

```python
months=["jan","feb","mar","april","may","june","july","august","sep","oct","nov","dec"]
year=["2011","2012","2013","2014","2015","2016","2017","2018","2019","2020","2021"]
plt.figure(figsize=(20,5))
for z in bop["new_year"].unique():
    a=bop.groupby(bop[bop["new_year"]==z]["new_month"])
    c=[]
    for f,b in a:
        c.append(b["Close"].mean())
    if(z==2021):
        for i in range(0,5):
                c.append(0)
    plt.title("bop monthly mean graph")
    plt.xlabel("months")
    plt.ylabel("mean value")   
    plt.legend(year,loc=2)
    plt.plot(months,c)
plt.show()

plt.figure(figsize=(20,5))
for z in abl["new_year"].unique():
    a=abl.groupby(abl[abl["new_year"]==z]["new_month"])
    c=[]
    for f,b in a:
        c.append(b["Close"].mean())
    if(z==2021):
        for i in range(0,5):
                c.append(0)
    plt.title("abl monthly mean graph")
    plt.xlabel("months")
    plt.ylabel("mean value")    
    plt.legend(year,loc=2)
    plt.plot(months,c)
plt.show()

plt.figure(figsize=(20,5))
for z in ubl["new_year"].unique():
    a=ubl.groupby(ubl[ubl["new_year"]==z]["new_month"])
    c=[]
    for f,b in a:
        c.append(b["Close"].mean())
    if(z==2021):
        for i in range(0,5):
                c.append(0)
    plt.title("ubl monthly mean graph")
    plt.xlabel("months")
    plt.ylabel("mean value")
    plt.legend(year,loc=2)
    plt.plot(months,c)
plt.show()

plt.figure(figsize=(20,5))
for z in hbl["new_year"].unique():
    a=hbl.groupby(hbl[hbl["new_year"]==z]["new_month"])
    c=[]
    for f,b in a:
        c.append(b["Close"].mean())
    if(z==2021):
        for i in range(0,5):
                c.append(0)
    plt.title("hbl monthly mean graph")
    plt.xlabel("months")
    plt.ylabel("mean value")
    plt.legend(year,loc=2)
    plt.plot(months,c)
plt.show()
```

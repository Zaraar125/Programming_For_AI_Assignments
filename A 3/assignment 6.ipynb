### Question 1 Part b  (10 marks)

Clean the Reviews using NLTK,RE. Remove Punctuations, Smilies and other garbage using any library. 

Apply Vader Sentiment to find the sentiment of the review, and append it in a column named "Sentiment".

You can get help of Vader sentiment from this; https://towardsdatascience.com/sentimental-analysis-using-vader-a3415fef7664

```python
# import nltk
# nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
from nltk.tokenize import word_tokenize,RegexpTokenizer
Names=df["Names"]
Review=df["Review Text"]
new_names=[]
lowered=[]
new_review=[]
for i in range(len(names)):
    new_names.append(Names[i].replace("Verified Purchase","").replace("by",""))
    lowered.append(Review[i].lower())
tokenizer=RegexpTokenizer(r"\w+")
for i in lowered:
    new_review.append(tokenizer.tokenize(i))
a_new_review=[]
for i in new_review:
    a=" ".join(str(read) for read in i)
    a_new_review.append(a)
    a=""
```

```python
new_df=pd.DataFrame((new_names,a_new_review))
new_df=new_df.T
```

```python
new_df.rename({0:"Names",1:"Review Text"},axis=1,inplace=True)
```

```python
new_df
```

```python
sentiments=[]
checker= SentimentIntensityAnalyzer()

to_check=new_df["Review Text"]

for i in to_check:
    sentiments.append(checker.polarity_scores(i))
```

```python
sent=[]
for i in sentiments:
    for n in i.keys():
        if n=="compound":
            if i[n]>0:
                sent.append("pos")
            else:
                sent.append("neg")
    
```

```python
new_df["Sentiments"]=sent
```

```python
check=-1
for i in range(0,80):
    
    driver=webdriver.Chrome(path)
    driver.get(url)
    time.sleep(8)
    driver.maximize_window()
    search=driver.find_element(By.ID,"q")
    search.send_keys(query)
    search.send_keys(Keys.RETURN)
    time.sleep(5)
      
    
  
    if i>40:
         driver.execute_script("window.scrollBy(0,1900)","")
         check=check+1
         l=driver.find_element(By.XPATH,"//*[@id='root']/div/div[2]/div/div/div[1]/div[3]/div/ul/li[3]/a")
         l.click()
         time.sleep(10)
    
    mobile_box=driver.find_element(By.CLASS_NAME,"box--ujueT")
    
    #mobile_info=mobile_box.find_elements(By.TAG_NAME,"a")
    
    mobile_info=mobile_box.find_elements(By.CLASS_NAME,"title--wFj93")
    driver.execute_script("window.scrollBy(0,300)","")
    if i<40:
        searching=mobile_info[i]
    else :
        searching=mobile_info[check]
    try:
        searching.click()
    except:
        driver.execute_script("window.scrollBy(0,60)","")
        searching.click()
    #mobile_info[i].click()
    time.sleep(5)
    price=driver.find_element(By.CLASS_NAME,"pdp-product-price").text
    #price=p.find_element(By.CLASS_NAME," pdp-price pdp-price_type_normal pdp-price_color_orange pdp-price_size_xl").text
    name=driver.find_element(By.CLASS_NAME,"pdp-mod-product-badge-title").text
    time.sleep(10)
  #  driver.execute_script("window.scrollBy(0,1350)","")
   # time.sleep(7)
  
    element =driver.find_element(By.XPATH,"//*[@id='module_product_review']/div/div/div[1]")
    #element =driver.find_element(By.XPATH,"//*[@id='module_product_review']/div/div/div[1]/div[2]/div/div/div[1]/div[1]/span[1]")
    driver.execute_script("arguments[0].scrollIntoView(true);",element)
    
    time.sleep(10)
    #star=driver.find_element(By.XPATH,"//*[@id='module_product_review']/div/div/div[1]/div[2]/div/div/div[1]/div[1]/span[1]").text
    star=driver.find_element(By.CLASS_NAME,"summary")
    star=star.find_element(By.CLASS_NAME,"score-average").text
    rate=driver.find_element(By.CLASS_NAME,"count").text
    time.sleep(10)
    names.append(name)
    prices.append(price)
    stars.append(star)
    rating.append(rate)
    driver.close()
    time.sleep(7)
    print(i)
```

```python
path="chromedriver.exe"
driver=webdriver.Chrome(path)
url="https://www.daraz.pk/catalog/?from=filter&q=mobile%20covers"
driver.get(url)
driver.maximize_window()
time.sleep(1)
driver.execute_script("window.scrollBy(0,100)","")
time.sleep(2)
search=driver.find_element(By.XPATH,"//*[@id='root']/div/div[2]/div/div/div[2]/div/div[3]/div[2]/div/div")
time.sleep(1)
#search1=driver.find_elements(By.CLASS_NAME,"ant-checkbox")
store_name=[]
lis=[]
mall_type=[]
country_type=[]
avg_price=[]
seller_rating=[]
followers=[]
for i in range(0,11):
     search1=driver.find_elements(By.CLASS_NAME,"ant-checkbox")
     search1[i].click()
     time.sleep(2)
     listings=driver.find_element(By.XPATH,"//*[@id='root']/div/div[2]/div/div/div[1]/div[1]/div/div[1]/div/div").text
     mall=driver.find_element(By.XPATH,"//*[@id='root']/div/div[2]/div/div/div[2]/div/div[3]/div[2]").text
     box=driver.find_element(By.CLASS_NAME,"box--ujueT")
     z=0
     items=0
     rate=0
     for pp in box.find_elements(By.CLASS_NAME,"price--NVB62"):
            items+=1
            if "," in pp.text.split(" ")[1]:
                    z=z+int(pp.text.split(" ")[1].replace(",",""))
            else :
                z=z+int(pp.text.split(" ")[1])
     avg_price.append(int(z/items)) 
     beta=list(listings.split(" "))
     if mall!="Pakistan" and mall!="China":
        mall="Daraz Mall"
     else:
        mall="Private Mall"
     country=driver.find_element(By.XPATH,"//*[@id='root']/div/div[2]/div/div/div[2]/div/div[4]/div[2]").text
     if country=="-":
            country=driver.find_element(By.XPATH,"//*[@id='root']/div/div[2]/div/div/div[2]/div/div[3]/div[2]").text
     
     if "Pakistan" in country :
                country_type.append("Pakistan")
     else:
            country_type.append("China")
   
     #take1=driver.find_element(By.XPATH,"//*[@id='root'/div/div[2]/div/div/div[1]/div[3]/div[1]/div/div/div[2]/div[2]/a")
     take1=driver.find_element(By.CLASS_NAME,"title--wFj93")
     take1.click()
     time.sleep(4)
     take2=driver.find_element(By.CLASS_NAME,"seller-link")
     time.sleep(5)
     try:
          take2.click()
     except:
         driver.execute_script("window.scrollBy(0,60)")
         take2.click()
     time.sleep(3)
     seller_rating.append(driver.find_element(By.XPATH,"//*[@id='pi-component-container']/div/div[2]/div/div/div/div[2]/div[1]/div/div/div/div[2]/div[3]").text[1])
     followers.append(driver.find_element(By.XPATH,"//*[@id='pi-component-container']/div/div[2]/div/div/div/div[2]/div[1]/div/div/div/div[2]/div[2]/div").text)
     lis.append(beta[0])
     mall_type.append(mall)
     store_name.append(beta[-1])
     time.sleep(3)
     driver.back()
     time.sleep(3)
     driver.back()
     time.sleep(4)
     #search1[i].click()
     driver.back()
     time.sleep(7)
f=[i.replace("Followers","") for i in followers]
driver.close()
```

```python
df=pd.DataFrame(columns=["Store Name","Daraz Mall/ Private","Avg Price","Listings","Seller Rating","Country","Followers"])
```

```python
df["Store Name"]=store_name
df["Daraz Mall/ Private"]=mall_type
df["Avg Price"]=avg_price
df["Listings"]=lis
df["Seller Rating"]=seller_rating
df["Country"]=country_type
df["Followers"]=f
```
